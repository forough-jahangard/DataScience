```python
data_new=data[(data['akharin takhor']<15)]
```

```python
data_new=data[(data['akharin takhor']<15)]
```

```python
data_new=data_new[data_new['max_takhor']<60]
```

```python
data_new.describe()
```

```python
data=pd.get_dummies(data_new)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
dataScale=scaler.fit_transform(data)
```

```python
from sklearn.cluster import KMeans


SSE = []

for i in range(2,11):
    
    km=KMeans(n_clusters=i,max_iter=200,random_state=42)
    
    km.fit(dataScale)
    
    SSE.append(km.inertia_)
    
    
SSE  
```

```python
plt.plot(range(2,11),SSE)

plt.xlabel('No of Clusters')

plt.ylabel('SSE')

plt.show()
```

```python
kmeans = KMeans(n_clusters=5,max_iter=200,random_state=42)

kmeans.fit(dataScale)
```

```python
cluster=kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
dataCluster=data.assign(cluterlabel=cluster)

dataCluster
```
